```python
# Import af biblioteker
from bs4 import BeautifulSoup
import re
```

```python
# indlæs xml filen
with open(r'C:\Users\lakj\Desktop\DH Teaterhistorie\Holbergs_comedies\1723_kandstøber.xml', 'r', encoding='utf-8-sig') as f:
    file = f.read() 

soup = BeautifulSoup(file, 'xml')
```

```python
# xml soup
print (soup.prettify())
```

```python
# tags i filen
tags = [tag.name for tag in soup.find_all(True)]
print (set(tags))
```

```python
# akterne
act1 = soup.find(id="act1")
act2 = soup.find(id="act2")
act3 = soup.find(id="act3")
act4 = soup.find(id="act4")
act5 = soup.find(id="act5")
```

```python
# transformer teksten fra xml til ensartet tekst
def get_text_out(in_data):
    import re
    act_raw = in_data.get_text()\
    #.replace('.', '').replace(';', '')\
    #.replace(',', '').replace(':', '').replace('!', '').replace('?', '')
    
    act_cl = ' '.join(re.findall('\S+', act_raw)).lower()
    return act_cl

act1_cl = get_text_out(act1)
act2_cl = get_text_out(act2)
act3_cl = get_text_out(act3)
act4_cl = get_text_out(act4)
act5_cl = get_text_out(act5)
```

```python
# gem renset tekst
def write_file(filename, string):
    of = open(filename+'.txt', 'w')
    of.write(string)
    
write_file('act1_cl', act1_cl)
write_file('act2_cl', act2_cl)
write_file('act3_cl', act3_cl)
write_file('act4_cl', act4_cl)
write_file('act5_cl', act5_cl)
```
